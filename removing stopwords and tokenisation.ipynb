```python
import pandas as pd
```

```python
df1=open("finalset data.csv")

df=pd.read_csv(df1)
df.head()
```

```python
test = pd.DataFrame(df)
test.columns = ["topics","volumes","volumes-href","papers"]
test["papers"] = test["papers"].str.lower().str.split()
```

```python
print(test["papers"])
```

```python
test['papers'].apply(lambda x: [item for item in x if item not in s])
```
